```python
def read_from_file(filename, width, start, end, skip):
    file = open(filename)
    content = file.readlines()
    
    boxen = []
    cell = [0]*width

    for line in range(start, end, skip):
        this_content = content[line]
        for i in range(width):
            cell[i] = content[line + i]
            a = re.findall("\d+\.\d+", cell[i])
            boxen.append(a)
        
    return boxen
```
